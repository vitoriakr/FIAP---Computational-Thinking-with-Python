Métodos de listas

```python
#.append() -> adiciona um elemento no final da lista
filmes = ['Matrix', 'Interestellar']
filmes.append('O Jogo da Imitação')
print(filmes)
```

```python
#copy -> copia a lista para outra variável
filmes = ['Matrix', 'Interestellar']
lista_nova = filmes.copy()
lista_nova.append('O Jogo da Imitação')
print(f'Lista original: {filmes}')
print(f'Lista nova: {lista_nova}')
```

```python
#.insert() -> adiciona um elemento em uma posição específica da lista
filmes.insert(1, 'Inception')
print(filmes)
```

REMOÇÃO DE ELEMENTOS

```python
#.remove() -> remove o item pelo primeiro que encontrar
filmes.remove('Inception')
print(filmes)
```

```python
#.pop() -> remove o item pelo índice
filmes.pop(0)
print(filmes)
```

Organização e ordenação

```python
#.sort() -> ordena a lista
idades = [10, 50, 2, 60]
idades.sort()
print(idades)
print(sorted(idades, reverse=True))
```

```python
series = [
    'Black Mirror',
    'The office',
    'Yellojackets',
    'Stranger Things',
]
```

```python
series.sort()
print(series)
```

```python
#.index() -> retorna o índice do primeiro elemento encontrado
print(series.index('The office'))
```

```python
posicao = series.index('The office')
print(f'O índice de The office é: {posicao}')
```

```python
#count -> conta quantas vezes um elemento aparece na lista
series.append('The office')
print(series.count('The office'))
```
