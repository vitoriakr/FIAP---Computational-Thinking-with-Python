TIPOS DE DADOS


```python
# String (texto)

nome = "João"
print(nome)
```

```python
# Inteiro

idade = 25
print(idade)
```

```python
# Float (número decimal)

altura = 1.75
print(altura)
```

```python
# Boolean (verdadeiro ou falso)

estudante = True
print(estudante)
```

```python
nome = "Ana"
idade = 20
altura = 1.68
estudante = True

print(type(nome))
print(type(idade))
print(type(altura))
print(type(estudante))
```
