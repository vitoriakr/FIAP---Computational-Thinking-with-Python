```python
idade = 18

# jeito 1 (separando com vírgula)
print("A minha idade é:", idade)

# jeito 2 (usando concatenação)
print("A minha idade é " + str(idade))

# jeito 3 (usando format)
print("A minha idade é {}".format(idade))

# jeito 4 (f-string - o mais usado hoje)
print(f"A minha idade é {idade}")
```
