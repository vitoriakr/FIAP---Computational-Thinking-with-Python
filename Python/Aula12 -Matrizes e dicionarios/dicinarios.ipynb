```python
#dicionario de um produto
produto ={
    'nome':'teclado',
    'preco': 150.0,
    'estoque':20
}
```

```python
type(produto)
print(produto['nome'])
```

**inserindo e atualizando elementos**

```python
produto['cor'] = 'cinza'
print(produto)
```

```python
produto['preco'] = 135.0
print(produto)
```

**removendo elementos**

```python
usuario = {
    'id':'707',
    'nome':'nana',
    'email':'nana@gmail.com',
    'status': 'ativo',
    'adimin': False
}
```

```python
#método pop()
email_removido = usuario.pop('email')
print(email_removido)
print(usuario)
```

```python
#comando del
del usuario['status']
print(usuario)
```

```python
#método clear()
usuario.clear()
print(usuario)
```

```python
#método keys, value e item
produto ={
    'nome':'teclado',
    'preco': 150.0,
    'estoque':20
}
print(produto.keys())
print(produto.values())
print(produto.items())
```

```python
# For - Chaves
for chave in produto.keys():
    print(chave)
```

```python
# For - Values
for valor in produto.values():
    print(valor)
```

```python
# For - Items
for item in produto.items():
    print(item)
```

```python
# For - Items
for chave, valor in produto.items():
    print(f"{chave}, {valor}")
```

```python
#desafio gerenciamento de estoque
frutas ={
    'Maça': 5,
    'Banana':3,
}
print(frutas)
frutas['Pera'] = 4
print(frutas)
```

```python
#método update
frutas.update({'Uva':8, 'Laranja':2})
print(frutas)
```
