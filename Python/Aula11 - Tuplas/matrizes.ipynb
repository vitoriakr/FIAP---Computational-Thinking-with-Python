```python
#criando matriz 3x3
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
matriz
```

```python
#criando matriz a partir de input
matriz=[]

for i in range(3):
    linha = []
    
    for j in range(3):
        valor = int(input(f'digite o valor para a posição [{i}][{j}]'))
        linha.append(valor)
        
    matriz.append(linha)

print(matriz)
```

```python
#criar matriz com numeros aleatorios -> 3x5
import random
matriz =[]
for i in range(3):
    linha = []
    
    for j in range(5):
        n = random.randint(1, 50)
        linha.append(n)
    matriz.append(linha)
    
for linha in matriz:
    print(linha)
```
