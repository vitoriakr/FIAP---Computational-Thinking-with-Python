```python
# Lista de tuplas (cada tupla é um ponto fixo x, y)
rota_entregas = [
    (-23.5505, -46.6333),
    (-23.5589, -46.6611),
    (-23.5505, -46.6333), # Entrega duplicada no mesmo prédio
    (-23.5921, -46.6723)
]
```

```python
print(len(rota_entregas))
```

```python
ponto_especifico = (-23.5505, -46.6333)
print(rota_entregas.count(ponto_especifico))
```

```python
for destino in rota_entregas:
    print(f'Destino: {destino}')
```
