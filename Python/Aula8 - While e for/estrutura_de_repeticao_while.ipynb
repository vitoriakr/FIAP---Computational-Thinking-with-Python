```python

# Desafio das médias

total_numeros = int(input("Quantas notas você quer inserir?"))

# Variáveis de controle
soma = 0
contador = 0

while contador < total_numeros:
    valor = float(input(f"Digite o {contador +1}º número: "))

#Adicionar o número digitado a nossa soma acumulada
    soma += valor

#Incrementar o contador para avançar para a próxima repetição
    contador +=1

media = soma / total_numeros
print(f"A média final é {media:.2f}")
```

```python
#jogo da adivinhação
from random import randint
numero_secreto = randint(1, 100)
tentativas = 0
while True:
    chute = int(input("Digite um número entre 1 e 100: "))
    tentativas += 1
    
    if chute < numero_secreto:
        print("Tente um número maior.")
    elif chute > numero_secreto:
        print("Tente um número menor.")
    else:
        print(f"Parabéns! Você acertou o número {numero_secreto} em {tentativas} tentativas!")
        break
```

```python
"""
desafio do orçamento
entendendo o while com um novo desafio
"""

orcamento = float(input("Digite o valor do orçamento total: R$ "))
total_gasto = 0
# O loop vai veridicar : o que eu já gastei ainda cabe no meu orçamento?
while total_gasto < orcamento:
    gasto = float(input(f"Digite o valor do gasto: R$ {total_gasto:.2f}: "))
    total_gasto += gasto
    
#quando o taotal gasto for igual ou maior que o orçamento, o loop para e mostra o total gasto
excedente = total_gasto - orcamento
print(f"Você excedeu o orçamento em R$ {excedente:.2f}")
```

```python
#DESAFIO: A VIAGEM DE CARRO
distancia_percorrida = 0
while distancia_percorrida < 100:
    km= float(input("quantos quilômetros você percorreu nessa última hora? km"))
    distancia_percorrida = km + distancia_percorrida
    print(f"Você percorreu {distancia_percorrida} km de 100 km até agora.")
    if distancia_percorrida == 100:
        print("Parabéns! Você chegou ao seu destino!")
    if distancia_percorrida > 100:
        print("Você ultrapassou a distância necessária para chegar ao destino!")
    
```

```python
# comando break
"""
    Contexto: 
    imagine que voce esta buscando um numero especifico em uma sequencia, assim que encontrar, não é motivo para continuar procurando.
""" 
numero_procurado = 5
tentativas =  1
while tentativas <= 10:
    if tentativas == numero_procurado:
        print(f"Número {numero_procurado} encontrado na tentativa {tentativas}!")
        break   
    print(f"Tentativa {tentativas}: número {tentativas} não é o número procurado.")
    tentativas += 1
```

```python
#carrinho de compras
total_compra = 0
while True:
    preco = float(input("Digite o preço do item (ou 0 para finalizar): R$ "))
    if preco == 0:
        break
    total_compra += preco

print(f"Total da compra: R$ {total_compra:.2f}")
```

```python
#comando continue
numero = 0
while numero < 10:
    numero += 1
    if numero == 5:
        print("Número 5 encontrado, pulando para a próxima iteração.")
        continue
    print(f"Número atual: {numero}")
        
```

```python
#playlist
musica_atual = 1
print('---PLAYER VIRTUAL---')
while musica_atual <= 10:
    print(f"Tocando música {musica_atual}...")
    musica_atual += 1
    #interação com o usuário
    acao = input("Digite 'o' para ouvir, 's' para sair ou 'p' para próxima música: ").lower()
    if acao == 's':
        print("Saindo do player. Até a próxima!")
        break
    elif acao == 'p':
        print(f"Pulando música {musica_atual} para a próxima música...")
        musica_atual += 1
        continue
    elif acao == 'o':
        print(f"Você está ouvindo a música {musica_atual - 1}.")
    else:
        print("Comando inválido. Por favor, digite 'o', 's' ou 'p'.")
print(f'voce ouviu a {musica_atual} inteira')
if musica_atual > 10:
    print("Você chegou ao final da playlist!")
else:    print(f"playlist encerrada na música {musica_atual}.")
```

```python
num_products = 4
discount = 0.15
for i in range(1, num_products + 1):
    price = float(input(f'Digite o preço do produto {i}: '))
    if price > 100:
        final_price = price * (1 - discount)
        print(f'O preço final do produto {i} com desconto é: R${final_price:.2f}')
    else:
        print(f'O preço do produto {i} é: R${price:.2f} (sem desconto)')
    
    
```
