# Estruturas Condicionais

#### IF:

```python
nota = float(input('Digite uma nota: '))

print(f'Nota do aluno: {nota}')

if nota >= 7:
    print("Aprovado!")
```

### ELSE:

```python
nota = float(input('Digite uma nota: '))

print(f'Nota do aluno: {nota}')

if nota >= 7:
    print("Aprovado!")
else:
    print("Reprovado!")
```

### ELIF:

```python
nota = float(input('Digite uma nota: '))

print(f'Nota do aluno: {nota}')

if nota >= 8:
    print("Aprovado Satisfatoramente")
elif nota == 7:
    print("Aprovado!")
else:
    print("Reprovado!")
```

```python
# Nós também podemos combinar várias condições a serem testadas
nota1 = float(input('Nota da primeira avaliação: '))
nota2 = float(input('Nota da segunda avaliação: '))
media = (nota1 + nota2) / 2

faltas = int(input('Quantidade de faltas no semestre: '))

if media >= 7 and faltas <= 10:
    print("Aprovado!")
elif media >= 7 and faltas > 10:
    print("Reprovado! por faltas")
else:
    print("Reprovado! media insuficiente")
```

```python
# Valor a pagar de uma compra no supermercado
tipo_user = input("Digite o tipo de usuário (com clube/sem clube): ").lower()
if tipo_user == "com clube":
    valor = 47.50
    print(f'valor a pagar R$: {valor:.2f} ')#formata o valor em 2 casas
elif tipo_user == "sem clube":
    valor = 50.0
    print(f'valor a pagar R$: {valor:.2f} ')
else:
    print('Não foi possível identificar o tipo de usuário')
```

### comandos úteis para manunipular strings
* len(): Retorna o tamanho(comprimento) de uma string

* in: Verifica se um pedaço de texto existe dentro de outro

* startwith(): Se a string começa com um determinado texto

* endswich(): Se a string termina com um determinado texto

* strip():Remove espaços em branco do inicio e fim da string

* upper():Transforma o testo em letras maiusculas

* lower(): Retorna a string em letras minusculas

* replace():Substitui uma substring por outra

### Aplicando esses conceitos a condicionais

```python
username = input('crie seu usuário: ')

if len(username) < 5:
    print('ERRO: o nome do usuário tem que ter pelo menos 5 caracteres')
elif len(username) > 20 :
    print('ERRO:Nome do usuário muito longo')
else:
    print(f'{username}: Nome disponível')
```

```python
#Exemplo: sistema de upload de Avatar
arquivo = input('Nome do arquivo: ').lower().strip()
if arquivo.endswith('.jpd') or arquivo.endswith('.png'):
    print('Upload permitido! O arquivo é uma imagem')
else:
    print('ERRO: apenas arquivos tipo .jpg ou .png são permitidos')
```

```python
raw_price = input('Valor do produto: ')

adjusted_price = raw_price.replace(',','.')
final_price = float(adjusted_price)
if final_price > 100:
    final_price = final_price * 0.9# ou final_price *= 0.9
    print(f'Desconto aplicado. Valor final: R${final_price}')
else:
    print(f'Valor final: R${final_price}')
```

```python
#DESAFIO:
valor_produto = int(input('Valor do produto: '))
categoria_produto = str(input('Categoria do produto(eletronicos, limpeza ou cozinha): '))
cupom_code = input('Cupom do produto: ')
if categoria_produto == 'eletronicos' and cupom_code == 'TECH20':
    print(f'Categoria do produto: {categoria_produto} tem o desconto de 20% com o cupom (TECH20). Valor final:R${valor_produto - (valor_produto * 0.20) }')
elif categoria_produto == 'limpeza' or 'cozinha' and cupom_code == 'CASA10':
    print(f'Categoria do produto: {categoria_produto} tem o desconto de 10% com o cupom (CASA10). Valor final :R${valor_produto - (valor_produto * 0.10 )}')
else:
    print(f'Categoria do produto: {categoria_produto} não tem desconto  Valor:R${valor_produto}')
```
