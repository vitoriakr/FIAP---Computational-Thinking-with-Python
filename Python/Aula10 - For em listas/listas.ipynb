TRATAMENTO DE DADOS

```python
#ETL - extract, transform, load - processo de extrair dados de uma fonte, transformar esses dados para um formato adequado e carregá-los em um sistema de destino, como um banco de dados ou um data warehouse.
lista_ceps_entrada = [
    '01001-000',
    '2004001',
    '70040-010',
    '098508.098',
    'Não Sei'
]
lista_ceps_normalizados = []

for registro_cep in lista_ceps_entrada:
    cep_formatado = registro_cep.strip().replace('-', '').replace('.', '')
    lista_ceps_normalizados.append(cep_formatado)
    print(f'Processado: {registro_cep} -> {cep_formatado}')

print(lista_ceps_normalizados)
```

```python
lista_ceps_entrada = [
    '01001-000',
    ' 2004001',
    '70040-010',
    '098508.098',
    'Não Sei'
]
lista_ceps_validados = []
lista_ceps_rejeitados = []

for registro in lista_ceps_entrada:
    #limpeza de dados
    cep_limpo = registro.strip().replace('-', '').replace('.', '')
    #validação de dados
    if len(cep_limpo) == 8 and cep_limpo.isdigit():
        lista_ceps_validados.append(cep_limpo)
    else:
        lista_ceps_rejeitados.append(registro)
print(f'CEPs Validados: {lista_ceps_validados}')
print(f'CEPs Rejeitados: {lista_ceps_rejeitados}')
```
