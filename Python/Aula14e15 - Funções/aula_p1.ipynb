## Desafio: Frete Inteligente

Contexto: Sua loja virtual oferece frete grátis para compras acima de um certo valor. Para compras menores, é cobrada uma taxa fixa. Você precisa criar uma função que receba o valor da compra e decida qual o valor final que o cliente deve pagar.

O que deve ser feito:

Crie uma função chamada calculate_shipping.

Essa função deve receber um parâmetro: purchase_amount.

Dentro da função:

Se o valor da compra for maior ou igual a 200 reais, o frete é R$ 0,00.

Se for menor que 200 reais, o frete é uma taxa fixa de R$ 20,00.

A função deve retornar (return) apenas o valor do frete.

No programa principal:

Peça ao usuário o valor total do carrinho de compras (via input).

Chame a função para descobrir o valor do frete.

Calcule o total final (valor da compra + frete).

Exiba os detalhes para o usuário: print(f"\nResumo do Pedido:") print(f"Compra: {}") print(f"Frete: {}") print(f"Total: {}")

```python
def calculate_shipping(purchase_amount):
    if purchase_amount >= 200.0:
        return 0
    else:
        return 20.00
#csrrinho
cart_total = float(input('Digite o preço total da sua compra(reais): '))
#frete
shiping_cost =calculate_shipping(cart_total)

final_total= cart_total+shiping_cost

print(f"\nResumo do Pedido:") 
print(f"Compra: {cart_total}") 
print(f"Frete: {shiping_cost}") 
print(f"Total: {final_total}")
```

###Desafio Prático: Currency Converter

Você está desenvolvendo o módulo financeiro de um sistema internacional. Todos os valores no seu banco de dados devem ser armazenados em Dólar (USD). O usuário pode inserir valores em Real (BRL) ou Euro (EUR). Sua missão é criar uma função que trate o texto digitado pelo usuário, converta para o câmbio correto e retorne o valor pronto para ser salvo.


O que a função (convert_to_usd) deve fazer:
Parâmetros: amount (valor) e currency_type (a moeda de origem).


Substituir vírgulas por pontos (para que o Python entenda como número decimal).


Remover qualquer caractere que não seja número ou ponto.
Converter o texto limpo para o tipo decimal (float).

```python
def convert_to_usd(amount, currency_type):
    clean_amount = amount.replace(',','.')
    value = float(clean_amount)
    
    if currency_type.upper() == 'BRL':
        usd_value = value/5
    elif currency_type.upper() =='EUR':
        usd_value = value*1.10
    else:
        usd_value = value
    return usd_value

print('----------CURRENCY CONVERTER TOOL-----------')
user_value = input('Coloque o valor a ser convertido: ')
user_currency_type = input('Moeda origem(BRL/EUR)')

final_price_usd = convert_to_usd(user_value, user_currency_type)

print(f'\n O valor final a ser salvo no db: ${final_price_usd} ')
```
