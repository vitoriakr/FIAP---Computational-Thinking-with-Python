```python
#segundo jeito de fazer
import re

def convert_to_usd(amount, currency_type):
    clean_amount = re.sub(r'[^0-9,.]','', amount)
    clean_amount = clean_amount.replace(",",".")
    
    value = float(clean_amount)
    
    currency = currency_type.upper().strip()
    
    if currency == 'BRL':
        usd_value = value/5
    elif currency =='EUR':
        usd_value = value*1.10
    else:
        usd_value = value
        
    return usd_value

print('----------CURRENCY CONVERTER TOOL-----------')
user_value = input('Coloque o valor a ser convertido: ')
user_currency = input('Moeda origem(BRL/EUR)')

final_price_usd = convert_to_usd(user_value, user_currency)

print(f'\n O valor final a ser salvo no db: ${final_price_usd} (dolar) ')
```
